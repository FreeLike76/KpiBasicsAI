# Imports

```python
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
```

# Functions

```python
def mfx_info(x, mfx):
    plt.figure(figsize=(16, 9))
    plt.plot(x, mfx, label="mfx")
    plt.legend()
    plt.show()
```

```python
def mfxs_info(x, mfxs):
    plt.figure(figsize=(16, 9))
    for name, mfx in mfxs.items():
        plt.plot(x, mfx, label=name)
    plt.legend()
    plt.show()
```

# Tasks
## Task 1

```python
x = np.linspace(0, 10, 1000)
```

```python
# triangle
# bottom1, top, bottom2
mfx = fuzz.trimf(x, [3, 5, 7])
mfx_info(x, mfx)
```

```python
# trapeze
# bottom1, top1, top2, bottom2
mfx = fuzz.trapmf(x, [2, 4, 7, 8])
mfx_info(x, mfx)
```

## Task 2

```python
# gaussian dist
# mean, sigma
mfx = fuzz.gaussmf(x, 5, 1)
mfx_info(x, mfx)
```

```python
# two gaussians
# mean1, sigma1, mean2, sigma2
mfx = fuzz.gauss2mf(x, 3, 1, 5, 2)
mfx_info(x, mfx)
```

```python
# two gaussians
# mean1, sigma1, mean2, sigma2
mfx = fuzz.gauss2mf(x, 4, 1, 4, 2)
mfx_info(x, mfx)
```

## Task 3

```python
# bell
# topsize, slope, center
mfx = fuzz.gbellmf(x, 1, 2, 5)
mfx_info(x, mfx)
```

## Task 4

```python
# sigmoid
# center, slope
mfx = fuzz.sigmf(x, 5, 3)
mfx_info(x, mfx)
```

```python
# sigmoid
# center, slope
mfx = fuzz.sigmf(x, 5, -1)
mfx_info(x, mfx)
```

```python
# sigmoid subtraction
# center1, slope1, center2, slope2
mfx = fuzz.dsigmf(x, 2, 2, 8, 2)
mfx_info(x, mfx)
```

```python
# sigmoid subtraction
# center1, slope1, center2, slope2
mfx = fuzz.dsigmf(x, 2, 2, 8, 1)
mfx_info(x, mfx)
```

```python
# sigmoid multiplication
# center1, slope1, center2, clope2
mfx = fuzz.psigmf(x, 2, 10, 5, 1)
mfx_info(x, mfx)
```

## Task 5

```python
# Z function
# top, bottom
mfx = fuzz.zmf(x, 3, 5)
mfx_info(x, mfx)
```

```python
# PI function
# bottom1, top1, top2, bottom2
mfx = fuzz.pimf(x, 2, 4, 6, 9)
mfx_info(x, mfx)
```

```python
# S function
# bottom1, top1, top2, bottom2
mfx = fuzz.smf(x, 4, 6)
mfx_info(x, mfx)
```

## Task 6

```python
# minimax disjunction
mfx1 = fuzz.gaussmf(x, 6, 2)
mfx2 = fuzz.gaussmf(x, 4, 2)
mfx_result = np.maximum(mfx1, mfx2)
mfxs_info(x, {"mfx1": mfx1, "mfx2": mfx2, "result": mfx_result})
```

```python
# minimax conjunction
mfx1 = fuzz.gaussmf(x, 6, 1)
mfx2 = fuzz.gaussmf(x, 4, 1)
mfx_result = np.minimum(mfx1, mfx2)
mfxs_info(x, {"mfx1": mfx1, "mfx2": mfx2, "result": mfx_result})
```

## Task 7

```python
# probabilistic disjunction
mfx1 = fuzz.gaussmf(x, 6, 2)
mfx2 = fuzz.gaussmf(x, 4, 2)
mfx_result = mfx1 + mfx2 - mfx1 * mfx2
mfxs_info(x, {"mfx1": mfx1, "mfx2": mfx2, "result": mfx_result})
```

```python
# probabilistic conjunction
mfx1 = fuzz.gaussmf(x, 6, 2)
mfx2 = fuzz.gaussmf(x, 4, 2)
mfx_result = mfx1 * mfx2
mfxs_info(x, {"mfx1": mfx1, "mfx2": mfx2, "result": mfx_result})
```

## Task 8

```python
phone_charge = np.arange(0, 101, dtype=np.int32)
mfx = fuzz.sigmf(phone_charge, 35, 0.14)
not_mfx = 1 - mfx

plt.figure(figsize=(16, 9))
plt.plot(phone_charge, mfx, label="Phone is charged")
plt.plot(phone_charge, not_mfx, label="Phone is NOT charged")
plt.legend()
plt.show()
```
